```python
import pandas as pd #判斷正反面 https://www.youtube.com/watch?v=SyYIuVJ9BhQ
import re
import numpy as np
TB =pd.read_csv('https://raw.githubusercontent.com/SophonPlus/ChineseNlpCorpus/master/datasets/waimai_10k/waimai_10k.csv')
pattern =re.compile('.{1}')
# pattern.findall(TB.review[0]) 分成單字
mydata=[pattern.findall(s) for s in TB.review]
import tensorflow as tf
tokenizer=tf.keras.preprocessing.text.Tokenizer(filters='')
tokenizer.fit_on_texts(mydata)
#tokenizer.index_word 看單字的標記
mytxttensor=tokenizer.texts_to_sequences(mydata)
mytxttensor=tf.keras.preprocessing.sequence.pad_sequences(
    mytxttensor,    
    padding='post',
    truncating='post',
    maxlen=30)
numwords=len(tokenizer.index_word)
embedding_dim=300
from tensorflow.keras import layers
embedding_layer=layers.Embedding(numwords+1,embedding_dim)
#embedding_layer(tf.constant([13,30,1]))維度
rnn_c=tf.keras.Sequential()
rnn_c.add(layers.Embedding(numwords+1,embedding_dim))
rnn_c.add(layers.SimpleRNN(64))
rnn_c.add(layers.Dense(2,activation='softmax'))#2是最後輸出結果,softmax算機率
tf.keras.utils.plot_model(rnn_c,show_shapes=True)
rnn_c.compile(optimizer='adam',
             loss=tf.losses.sparse_categorical_crossentropy,
             metrics=['accuracy'])
#rnn_c.summary()顯示參數
rnn_c.fit(mytxttensor,TB.label,epochs=20)#匯入資料到模組
newsen=[pattern.findall('不好吃')]
newsen=tokenizer.texts_to_sequences(newsen)
newsen=tf.keras.preprocessing.sequence.pad_sequences(
    newsen,
    padding='post',
    truncating='post',
    maxlen=30
    )
#print(rnn_c.predict(newsen))
#print(np.argmax(rnn_c.predict(newsen),axis=-1))





rnn_c.save('./keras_model.h5') #存檔
print('ok')
```

```python
import pandas as pd
import re
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers
TB =pd.read_csv('./1111.csv', encoding = "Big5") #讀檔
pattern =re.compile('.{1}')                      #切割字元
mydata=[pattern.findall(s) for s in TB.review]   #切割字元


tokenizer=tf.keras.preprocessing.text.Tokenizer(filters='')
tokenizer.fit_on_texts(mydata)
#tokenizer.index_word                             #看單字的標記
mytxttensor=tokenizer.texts_to_sequences(mydata)
mytxttensor=tf.keras.preprocessing.sequence.pad_sequences(
    mytxttensor,    
    padding='post',
    truncating='post',
    maxlen=10)
numwords=len(tokenizer.index_word)
embedding_dim=300
embedding_layer=layers.Embedding(numwords+1,embedding_dim)
#embedding_layer(tf.constant([13,30,1]))#維度
rnn_c=tf.keras.Sequential()                     #先建一個空的model
rnn_c.add(layers.Embedding(numwords+1,embedding_dim))
rnn_c.add(layers.SimpleRNN(64))
rnn_c.add(layers.Dense(5,activation='softmax'))  #5是最後輸出結果,softmax算機率
#tf.keras.utils.plot_model(rnn_c,show_shapes=True)  #劃出模型
rnn_c.compile(optimizer='adam',
             loss=tf.losses.sparse_categorical_crossentropy,
             metrics=['accuracy'])
#rnn_c.summary()#顯示參數
rnn_c.fit(mytxttensor,TB.label,epochs=20,      )#匯入資料到模組

newsen=[pattern.findall('五奈米續強，台積電Q4營收估季增近中位數百分比，毛利率51～53%')] #判斷字串
newsen=tokenizer.texts_to_sequences(newsen)
newsen=tf.keras.preprocessing.sequence.pad_sequences(
    newsen,
    padding='post',
    truncating='post',
    maxlen=30
    )

print('百分之',round(rnn_c.predict(newsen)[0][0]*100),'為極好 , ' +
      '百分之',round(rnn_c.predict(newsen)[0][1]*100),'為好  , '  +
      '百分之',round(rnn_c.predict(newsen)[0][2]*100),'為普通 , ' +
      '百分之',round(rnn_c.predict(newsen)[0][3]*100),'為不好 , ' +
      '百分之',round(rnn_c.predict(newsen)[0][4]*100),'為非常不好'
     )
if (np.argmax(rnn_c.predict(newsen),axis=-1)) == 0 :
    print('極好')
if (np.argmax(rnn_c.predict(newsen),axis=-1)) == 1 :
    print('好')
if (np.argmax(rnn_c.predict(newsen),axis=-1)) == 2 :
    print('普通')
if (np.argmax(rnn_c.predict(newsen),axis=-1)) == 3 :
    print('不好')
if (np.argmax(rnn_c.predict(newsen),axis=-1)) == 3 :
    print('非常不好')

```
